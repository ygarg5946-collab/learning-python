```python
#Now the main game begins the hangman problem
stages=[''' +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========''',
'''
 +---+
  |   |
  O   |
 /|\  |
      |
      |
========= ''',
'''
 +---+
  |   |
  O   |
 /|   |
      |
      |
=========''',
        '''
+---+
  |   |
  O   |
  |   |
      |
      |
=========''',
'''
 +---+
  |   |
  O   |
      |
      |
      |
=========''',
'''
 +---+
  |   |
      |
      |
      |
      |
=========''']
word_list=['Aardvark','baboon','camel']
import random
chosen_word=word_list[random.randint(0,len(word_list)-1)].lower()
print(f"The randomly chosen word is {chosen_word}")
x=len(chosen_word)
blank_list=[]
for i in range(x):
    blank_list.append("_")
lives=6
game_over=False
while not game_over:
    guess_word=input("Guess The alphabet: ")
    for i in range(len(chosen_word)):
        if chosen_word[i]==guess_word:
            blank_list[i]=guess_word
            print(stages[lives-1])
        else:
            lives-=1
            print(stages[lives-1])
if "_" not in blank_list:
    game_over=True

```

```python
#program to check whether the given number is prime or not
number=int(input("Enter the number:"))
for i in range(2,int(number/2)):
    if i==2:
        is_prime=True
    is_prime=True
    if number % i == 0:
        is_prime=False
        break
if is_prime==True:
    print(f"The number {number} is prime")
else:
    print(f"The number {number} is not prime")
```

```python
#write a program to check whether the given number is armstrong or not
number=input("Enter the number:")
x=len(number)
SUM=0
for i in number:
    SUM+=int(i)**x
if SUM==int(number):
    print(f"The given number {number} is armstrong")
else:
    print(f"The given number {number} is not armstrong")
```

```python
#Write a program to print armstrong numbers in an given interval
start=int(input("Enter the start number:"))
end=int(input("Enter the end number:"))
blank_list=[]
for i in range(start,end+1):
    x=len(str(i))
    SUM=0
    for j in str(i):
        SUM+=int(j)**x
    if SUM==i:
        blank_list.append(i)
print("The Armstrong numbers in the given range are:")
for k in blank_list:
    print(k)
```

```python
#now pyramid
rows=int(input("Enter the number of rows:"))
for i in range(1,rows+1):
    print(" "*(rows-i),end="")
    print("* "*i)
```

```python
#Bubble Sorting
LIST=input("Enter the elements in the LIST:").split()
LIST=[int(i) for i in LIST]
x=len(LIST)
for i in range(x-1):
    for j in range(x-i-1):
        if LIST[j]>LIST[j+1]:
            LIST[j],LIST[j+1]=LIST[j+1],LIST[j]
print(LIST)
```

```python
#Selection sort now
LIST=input("Enter the numbers in the list:").split()
LIST=[int(i) for i in LIST]
x=len(LIST)
for i in range(x-1):
    MIN_IDX=i
    for j in range(i+1,x):
        if LIST[MIN_IDX]>LIST[j]:
            MIN_IDX=j
    LIST[MIN_IDX],LIST[i]=LIST[i],LIST[MIN_IDX]
print(LIST)
```
